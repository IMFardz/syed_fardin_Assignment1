```python
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
def func(c):
    values = np.zeros([100])
    for i in range(1, 100):
        values[i] = (values[i - 1])**2 + c
    return values
    
```

Clearly, c = 0 converges as shown below

```python
a = func(np.complex64(0))
plt.plot(a)
```

c = 0.5 does not converge

```python
b = func(np.complex64(0.5))
plt.plot(b)
```

The following lines tests values that are purely real:

```python
reals = np.arange(-2, 2, step = 0.5)
c = func(reals[0])
plt.plot(c)
print(c)
```

```python
reals = np.arange(-2, 2, step = 0.5)
c = func(reals[1])
plt.plot(c)
print(c)
```
